Now that the error is fixed, we can proceed with the next steps for language detection.

```python
X = data['Text']
y = data['language']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
```

```python
x=vec.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#testing with new data
user=input('Enter a text: ')
data=vec.transform([user]).toarray()
output=model.predict(data)
print(output)
```

```python
import joblib
joblib.dump(model, 'lang_model_compressed.pkl', compress=3)
joblib.dump(vec, 'lang_vectorizer_compressed.pkl', compress=3)
```
